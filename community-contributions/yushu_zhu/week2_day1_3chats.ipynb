````python
gpt_model_one= "gpt-4o-mini"
gpt_model_two = "gpt-5-mini"
gpt_model_three = "gpt-4-turbo"

gpt_system_one = "你是刘备"

gpt_system_two = "你是曹操"

gpt_system_three= "你是孙权"


gpt_messages_one = ["我想和诸位讨论一下赤壁之战的策略"]
gpt_messages_two= ["好的"]
gpt_messages_three = ["让我们开始讨论吧"]
````

````python
def call_gpt_one():
    messages = [{"role":"system", "content":gpt_system_one}]

    for one,two,three in zip(gpt_messages_one, gpt_messages_two, gpt_messages_three):
        messages.append({"role": "assistant", "content": one})
        messages.append({"role": "user", "content": two})
        messages.append({"role": "user", "content": three})

    response = openai.chat.completions.create(
        model = gpt_model_one,
        messages = messages
    )
    return response.choices[0].message.content.strip()

def call_gpt_two():
    messages = [{"role":"system", "content":gpt_system_two}]

    for one,two,three in zip(gpt_messages_one, gpt_messages_two, gpt_messages_three):
        messages.append({"role": "user", "content": one})
        messages.append({"role": "assistant", "content": two})
        messages.append({"role": "user", "content": three})
    
    messages.append({"role":"user", "content": gpt_messages_one[-1]})
    response = openai.chat.completions.create(
        model = gpt_model_two,
        messages = messages
    )
    return response.choices[0].message.content.strip()

def call_gpt_three():
    messages = [{"role":"system", "content":gpt_system_three}]

    for one,two,three in zip(gpt_messages_one, gpt_messages_two, gpt_messages_three):
        messages.append({"role": "user", "content": one})
        messages.append({"role": "user", "content": two})
        messages.append({"role": "assistant", "content": three})
    
    messages.append({"role":"user", "content": gpt_messages_one[-1]})
    messages.append({"role":"user", "content": gpt_messages_two[-1]})
    
    response = openai.chat.completions.create(
        model = gpt_model_three,
        messages = messages
    )
    return response.choices[0].message.content.strip()

````

````python
display(Markdown(f"### 刘备:\n{gpt_messages_one[0]}\n"))
display(Markdown(f"### 曹操:\n{gpt_messages_two[0]}\n"))
display(Markdown(f"### 孙权:\n{gpt_messages_three[0]}\n"))

for i in range(5):
    gpt_next = call_gpt_one()
    display(Markdown(f"### 刘备:\n{gpt_next}\n"))
    gpt_messages_one.append(gpt_next)

    gpt_next = call_gpt_two()
    display(Markdown(f"### 曹操:\n{gpt_next}\n"))
    gpt_messages_two.append(gpt_next)

    gpt_next = call_gpt_three()
    display(Markdown(f"### 孙权:\n{gpt_next}\n"))
    gpt_messages_three.append(gpt_next)
````
